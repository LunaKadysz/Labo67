```python
import numpy as np

import pandas as pd

from scipy import interpolate
from scipy import integrate
import scipy.constants as ctte

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 18})
import matplotlib.cm as cm  # importaci√≥n de mapa de colores
mycmap = cm.plasma

```

```python
path1 = r'capacidad_serie_2.csv'
df_serie = pd.read_csv (path1)

path2 = r'capacidad_paralelo.csv'
df_paralelo = pd.read_csv (path2)
```

```python
df_serie.columns
```

```python

fig,ax = plt.subplots(figsize=(7,6))

ax.set_title("Capacidad en Serie (F)")
# make a plot
ax.grid()
ax.errorbar(range(1,len(df_serie)+1), df_serie['C_gap'], yerr = df_serie['C_gap_err'],color='orange', fmt='o')

ax.set_xlabel(r"Factor de llenado ($\nu$)", fontsize = 14)
ax.set_ylabel("En el gap [F]",color="orange",fontsize=15)
ax.tick_params(axis='both', labelsize=13)

ax2=ax.twinx()
ax2.tick_params(axis='both', labelsize=13)
ax.errorbar(range(1,len(df_serie)+1), df_serie['C_landau'], yerr = df_serie['C_landau_err'],color='green', fmt='o')
ax2.set_ylabel("Fuera del gap [F]",color="green",fontsize=15)
plt.show()



# paralelo

fig,ax = plt.subplots(figsize=(7,6))

ax.set_title("Capacidad en Paralelo(F)")
# make a plot
ax.grid()
ax.errorbar(range(1,len(df_paralelo)+1), df_paralelo['C_gap'], yerr = df_paralelo['C_gap_err'],color='orange', fmt='o')

ax.set_xlabel(r"Factor de llenado ($\nu$)", fontsize = 14)
ax.set_ylabel("En el gap [F]",color="orange",fontsize=15)
ax.tick_params(axis='both', labelsize=13)

ax2=ax.twinx()
ax2.tick_params(axis='both', labelsize=13)
ax.errorbar(range(1,len(df_paralelo)+1), df_paralelo['C_landau'], yerr = df_paralelo['C_landau_err'],color='green', fmt='o')
ax2.set_ylabel("Fuera del gap [F]",color="green",fontsize=15)
plt.show()
```
