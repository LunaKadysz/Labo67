```python
#descargo los modulos y librerias que se utilizan 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
data = {freq:{'ch1': [],'ch2': [], 't': []} for freq in ['1hz','10hz','100hz','1khz','10khz','50khz','100khz']}

for j,file in enumerate(os.listdir('data/')): #estoy diciendo que me recorra cada archivo de esa carpeta
    
    if os.path.splitext(file)[1] != 'npy':
        name = os.path.splitext(file)[0].split('_')
    
        chanel = name[0]
        freq = name[1]
        
        file_data = np.load(f'data/{chanel}_{freq}_data.csv.npy',allow_pickle=True).tolist()
        vdiv = float(file_data['V_div'])
        ofst = float(file_data['offset'])
        tdiv = float(file_data['T_div'])
        #N = int(file_data['n_meas'])
        if chanel == 'ch1':
            data[freq]['ch1'] = np.loadtxt(f"data/ch1_{freq}.csv",dtype=float)/32*float(vdiv)-float(ofst) #intr ricky
            N = len(data[freq]['ch1'])
            data[freq]['t'] = np.linspace(-tdiv*14/2,-tdiv*14/2+1,N) #intrucciones de ricky
        if chanel == 'ch2':
            data[freq]['ch2'] = np.loadtxt(f"data/ch2_{freq}.csv",dtype=float)/32*float(vdiv)-float(ofst) #intr ricky
    
```

# calculo overshoot 
$\frac{max - (mean-base)}{mean - base}$

```python
def ovsh(base,top,max_):
    salto = top - base
    ovsh = max_ - salto
    return ((max_-base) - (top-base)) / (top-base)

def damp(ovs):
    return -np.log(ovs)/(np.sqrt(np.pi**2 + (np.log(ovs))**2 ))

def t_s(tolerance,damp,frec):
    return np.log(tolerance * np.sqrt(1-damp**2) ) / (damp*frec)
  
```

```python
%matplotlib qt
t = data['1hz']['t']
in_ = data['1hz']['ch1']
out_ = data['1hz']['ch2']  
  
plt.plot(t, out_)
plt.xlim(0.49,0.5)
  
print("After 5 clicks :")
base,top,max_,t1,t2 = plt.ginput(5)
plt.show()
 
base = base[1]
top = top[1]
max_ = max_[1]
frec = 1/(t2[0] - t1[0])
ovs = ovsh(base,top,max_)
damp_  = damp(ovs)
ts = t_s(0.02,damp_,frec)
print(f' OVSH: {ovs}' )
print(f'Damping: {damp_}')
print(f'Settling time ts: {np.abs(ts)}')
```

# calculo damping

$\tau = \frac{-ln(OVSH)}{sqrt{\pi^2 + ln^2(OVSH)}}$
source: https://encyclopedia.pub/entry/29032

## settling time

$Ts = ln(tolerance * sqrt{1-damp^2} ) / (damp*frec)$
